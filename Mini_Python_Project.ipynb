```python
#open and reading text file
try:
    file=open('file.txt','r')
    cont=file.read()
    print("file content:",cont)
except:
    print("Exception occur")
```

```python
#using split function to convert a file into a list
try:
    contnt_list=cont.split(",")
    file.close()
    print("file List:",contnt_list)
except:
    print("An exception occured")
```

```python
#using map function
new_list=map(lambda x:x.upper(),contnt_list)
list(new_list)
```

```python
#using lambda function
def func1(s):
    try:
        if(len(s)%2==0):
            return s
    except:
        print("An error occured:")
if __name__=="__main__":
    new1_list=list(filter(func1,contnt_list))
    print(new1_list)
    
```

```python
cnt=new1_list
try:
    cnt.sort()
    cnt
except:
    print("An error occured")
```

```python
try:
    with open('file1.txt','w+') as f:
        for i in new1_list:
            f.write('%s\n'%i)
except:
    print("An error occured")

try:
    with open('file1.txt') as f:
        print(f.read())
    
except:
    print("An exception occured")
```
