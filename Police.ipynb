```python
def solve1(A,B):
    max1=0
    min1=0
    ans=0
    for i in range(len(A)):
        max1=max(max1,A[i])
    while(min1<=max1):
        mid=min1+(max1-min1)//2
        wood=0
        for j in range(len(A)):
            if(A[j]-mid>0):
                wood= wood +A[j]-mid
        if(wood>=B):
            min1=mid+1
            ans=max(ans,mid)
        else:
            mid -=1
    return ans
```
